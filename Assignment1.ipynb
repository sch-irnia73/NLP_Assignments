## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
import re
from collections import defaultdict, Counter
from typing import List

class BPETokenizer:

    def __init__(self, vocab_size: int) -> None:
        self.vocab_size = vocab_size
        self.vocab = None
        self.merges = []
    
    def get_stats(self, vocab: dict) -> dict:
        #Compute frequency statistics for character pairs in the vocabulary.
        pairs = defaultdict(int)
        for word, freq in vocab.items():
            symbols = word.split()
            for i in range(len(symbols) -  1):
                pairs[symbols[i], symbols[i +  1]] += freq
        return pairs
    
    def merge_vocab(self, pair: tuple, vocab: dict) -> dict:
        #Merge the most common pair of consecutive characters in the vocabulary.
        v_out = {}
        bigram = re.escape(' '.join(pair))
        p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
        for word in vocab:
            w_out = p.sub(''.join(pair), word)
            v_out[w_out] = vocab[word]
        return v_out
    
    def get_vocab(self, corpus: List[str]) -> dict:
        #Create a vocabulary from the given corpus.
        vocab = defaultdict(int)
        for line in corpus:
            for word in line.split():
                vocab[' '.join(list(word)) + ' </w>'] +=  1
        return vocab

    def train(self, corpus: List[str]) -> None:
        #Train the tokenizer with the given corpus.
        vocab = self.get_vocab(corpus)
        for _ in range(self.vocab_size):
            pairs = self.get_stats(vocab)
            best = max(pairs, key=pairs.get)
            vocab = self.merge_vocab(best, vocab)
            self.merges.append(best)
        self.vocab = vocab
        # print(self.vocab)

    def tokenize(self, text: str) -> List[str]:
        #Tokenize the given text using the trained vocabulary and merges.
        words = [list(word) + ['</w>'] for word in text.split()]

        for _ in range (2):
            for l, r in self.merges:
                for idx, word in enumerate(words):
                    new_word = []
                    i = 0
                    while i < len(word):
                        if i+1 <len(word) and word[i]==l and word[i+1]==r:
                            new_word.append(l+r)
                            i+=2
                        else:
                            new_word.append(word[i])
                            i+=1
                    words[idx] = new_word
        return [w.replace('</w>', '') for word in words for w in word]
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
# Your code here
import matplotlib.pyplot as plt

def read_files_into_list(filenames):
    combined_list = []
    for filename in filenames:
        with open(filename, 'r', encoding='utf-8') as file:
            for line in file:
                combined_list.append(line.strip())
    return combined_list

file_names = ['file1', 'file2', 'file3', 'file4', 'file5']

tweets = read_files_into_list(file_names)

# List of vocanulary sizes for the experiment
vocab_sizes = [1000,  750,  500,  250,  len(tweets)]
# print(list(tweets[0]))
# Dictionary for storing average lengths of tokenized tweets
avg_lengths = {}

for vocab_size in vocab_sizes:
    tokenizer = BPETokenizer(vocab_size)
    tokenizer.train(tweets)
    
    tokenized_tweets = [tokenizer.tokenize(tweet) for tweet in tweets]
    # print(vocab_size, tokenized_tweets)
    lengths = [len(tokens) for tokens in tokenized_tweets]
    avg_length = sum(lengths) / len(lengths)
    
    avg_lengths[vocab_size] = avg_length

# Plotting graph
plt.figure(figsize=(10,  5))
plt.grid(zorder=0)
plt.scatter(list(avg_lengths.keys()), list(avg_lengths.values()), color='blue')
plt.plot(list(avg_lengths.keys()), list(avg_lengths.values()), color='blue', zorder=3)
plt.xlabel('Vocab Size')
plt.ylabel('Average Length of Tokenized Tweet')
plt.title('Dependency of Average Length of Tokenized Tweet on Vocab Size')
plt.show()


# Preffered vocab size
preferred_vocab_size = min(avg_lengths, key=avg_lengths.get)
print(f"Preferred vocab size: {preferred_vocab_size}")
```

According to the analysis presented above, the most preferrable `vocab_size` appears to be 1000. This end is derived from the statement that a vocabulary length of 1000 yields the bottom average range of tokens per tweet, thereby minimizing the computational complexity related to processing the facts. Additionally, it's miles neither excessively small nor excessively large, as a consequence heading off potential drawbacks related to either severe.
